```python
# !pip install beautifulsoup4
# !pip install html5lib
```

```python
#finalmente una de los metodos mas interesantes, aunque uno de los mas complicados de ejecutar, el web scraping
#como su nombre lo indica, corresponde a extraer informacion de interes a partir del mismo codigo fuente de una pagina
#por lo que es usual que se requiera de diferentes algoritmos o metodos para cada objetivo de extraccion, haciendo de este metodo
#un conjunto de algoritmos no estandarizados, que suelen ser personalizados para cada caso, 

#para esta seccion veremos un pequeño vistazo de como se ejecutaria en sus versiones mas simples 

#inicialmente elegimos una pagina e implementamos la libreria request con su atributo get, que nos extraera el codigo fuente
#de la pagina
url = "https://www.ecologiaverde.com/las-ciudades-mas-contaminadas-del-mundo-6.html"
r = requests.get(url)
```

```python
r.content
```

```python
from bs4 import BeautifulSoup
import html5lib

#luego arreglaremos el codigo fuente, utilizando el formato html, el cual es el codigo fuente y estructura usual de
#las interfaces de las paginas web 
soup = BeautifulSoup(r.content,"html5lib")
```

```python
soup
```

```python
#Utilizamos la libreria soup, para identificar todos los indices de la pagina web, es decir todos los titulos
#por lo que extraeremos en forma de top, los paises mas contaminados, debido a la disposicion de esta pagina y su informacion
table = soup.find("div", attrs={"class":"indice"})
```

```python
table
```

```python
# finalmente iteramos a partir de todos los objetos vistos anteriormente con un for que recorra cada elemento
# en este caso, solo nos interesa los elementos encerrados por un </li> que es el comando html para declarar listas
#en este caso, una lista de paises contaminantes
quotes = []
for row in table.findAll("li"):
    quote = {}
    try:
        quote["Paises"] = row.contents[0].contents[0]
        quotes.append(quote)
    except TypeError:
        continue
```

```python
#Finalmente mostramos los resultados 
quotes
```
